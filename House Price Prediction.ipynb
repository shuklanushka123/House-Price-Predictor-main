# Outlier detection and removal

```python
data.describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby("Address"):
        m =np.mean(subdf.Price_sqft)
        
        st = np.std(subdf.Price_sqft)
        
        gen_df = subdf[(subdf.Price_sqft > (m-st)) & (subdf.Price_sqft <=(m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index = True)
    return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
def bedrooms_outlier_remover(df):
    exclude_indices = np.array([])
    for Addresss , Address_df in df.groupby("Address"):
        bedrooms_stats = {}
        for bedrooms , bedrooms_df in Address_df.groupby("Bedrooms"):
            bedrooms_stats[bedrooms] = {
                "mean" : np.mean(bedrooms_df.Price_sqft),
                "std" : np.std(bedrooms_df.Price_sqft),
                "count" : bedrooms_df.shape[0]
            }
        for bedrooms,bedrooms_df in Address_df.groupby("Bedrooms"):
            stats = bedrooms_stats.get(bedrooms-1)
            if stats and stats["count"] > 5:
                exclude_indices = np.append(exclude_indices , bedrooms_df[bedrooms_df.Price_sqft<(stats["mean"])].index.values)
    return df.drop(exclude_indices,axis = "index")
```

```python
data = bedrooms_outlier_remover(data)
```
